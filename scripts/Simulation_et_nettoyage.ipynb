```python
import pandas as pd
import numpy as np

# Nombre de lignes de données à générer
n = 10000

# Création de la série d'identifiants uniques des clients
cust_id = pd.Series(['CUST{}'.format(i) for i in range(1, n+1)])

# Génération de données aléatoires pour les variables numériques
balance = np.random.normal(5000, 1000, n)
balance_freq = np.random.uniform(0, 1, n)
purchases = np.random.normal(3000, 1000, n)
oneoff_purchases = np.random.normal(1000, 500, n)
installments_purchases = purchases - oneoff_purchases
cash_advance = np.random.normal(1000, 500, n)
purchases_freq = np.random.uniform(0, 1, n)
oneoff_purchases_freq = np.random.uniform(0, 1, n)
purchases_installments_freq = np.random.uniform(0, 1, n)
cash_advance_freq = np.random.uniform(0, 1, n)
cash_advance_trx = np.random.randint(0, 10, n)
purchases_trx = np.random.randint(0, 100, n)
credit_limit = np.random.normal(10000, 5000, n)
payments = np.random.normal(2000, 500, n)
minimum_payments = np.random.normal(500, 100, n)
prc_full_payment = np.random.uniform(0, 1, n)
tenure = np.random.randint(1, 12, n)

# Génération de données aléatoires pour les variables catégorielles
credit_card_type = pd.Series(np.random.choice(['Visa', 'MasterCard', 'American Express'], n, p=[0.4, 0.4, 0.2]))
account_type = pd.Series(np.random.choice(['Compte courant', 'Compte épargne'], n, p=[0.7, 0.3]))
income_category = pd.Series(np.random.choice(['Bas', 'Moyen', 'Élevé'], n, p=[0.4, 0.4, 0.2]))
type_of_eco_activity = pd.Series(np.random.choice(['Entreprises', 'Particuliers', 'Investisseurs'], n, p=[0.6, 0.3, 0.1]))

# Création d'un DataFrame pandas pour stocker les données
df = pd.DataFrame({
    'CUST_ID': cust_id,
    'BALANCE': balance,
    'BALANCE_FREQUENCY': balance_freq,
    'PURCHASES': purchases,
    'ONEOFF_PURCHASES': oneoff_purchases,
    'INSTALLMENTS_PURCHASES': installments_purchases,
    'CASH_ADVANCE': cash_advance,
    'PURCHASES_FREQUENCY': purchases_freq,
    'ONEOFF_PURCHASES_FREQUENCY': oneoff_purchases_freq,
    'PURCHASES_INSTALLMENTS_FREQUENCY': purchases_installments_freq,
    'CASH_ADVANCE_FREQUENCY': cash_advance_freq,
    'CASH_ADVANCE_TRX': cash_advance_trx,
    'PURCHASES_TRX': purchases_trx,
    'CREDIT_LIMIT': credit_limit,
    'PAYMENTS': payments,
    'MINIMUM_PAYMENTS': minimum_payments,
    'PRC_FULL_PAYMENT': prc_full_payment,
    'TENURE': tenure,
    'CREDIT_CARD_TYPE':credit_card_type,
    'ACCOUNT_TYPE':account_type,
    'INCOME_CATEGORY':income_category,
    'TYPE_OF_ECO_ACTIVITY':type_of_eco_activity})


# Enregistrement des données dans un fichier CSV avec les noms de colonnes
df.to_csv("../data/donnees_bancaires.csv", index=False)
```
